#### Plot the correlation matrix using seaborn.

```python
# your code here
corr = titanic.corr(method='pearson')
```

```python
corr
```

```python
plt.rcParams['figure.figsize']=(16, 10)
sns.set(style='ticks')    

mascara = np.triu(np.ones_like(corr, dtype=bool))   # genera una mascara

cmap = sns.color_palette("YlOrBr", as_cmap=True)   # paleta de colores


sns.heatmap(corr,
            mask = mascara, 
            cmap = cmap,
            vmax = 1,
            vmin = -1,
            center = 0,
            square = True,
            linewidth = 0.5,
            cbar_kws = {'shrink': 0.5},
            annot = True
           );
```
